```python
df_pass=df_premier[df_premier.eventId == 8]
```

```python
df_Fin_pass=df_pass.drop(['eventName','eventSec','id','matchId','matchPeriod','positions'],axis=1)
```

```python
df_Fin_pass.head()
```

```python
accu_pass = []
wrong_pass = []

for index,row in df_Fin_pass.iterrows():
    for value in row['tags'][-1].values():
        if value == 1801:
            accu_pass.append(1)
            wrong_pass.append(0)
        elif value == 1802:
            wrong_pass.append(1)
            accu_pass.append(0)
```

```python
df_Fin_pass['Accurate'] = accu_pass
df_Fin_pass['NotAccurate'] = wrong_pass
```
